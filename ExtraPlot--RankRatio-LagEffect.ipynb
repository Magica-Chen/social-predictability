# How many co-locators == social ties

```python
wp_top10 = pd.read_csv('final/extra/wp_common_top10_details.csv')
```

```python
def equality_ratio(df, denominator):
    egolist = list(df['userid_x'].unique())
    
    top10_SRN = df[df['category']=='social network']
    top10_NSCLN = df[df['category']=='non-social co-location network']
    
    top_denominator_SRN = top10_SRN[top10_SRN['Rank'] == denominator]
    
    container = []
    for ego in egolist:
        interim = top10_NSCLN[top10_NSCLN['userid_x'] == ego]
        value = top_denominator_SRN[top_denominator_SRN['userid_x'] == ego]['Pi_alters'].values
        
        interim['equality_ratio'] = value / interim['Pi_alters']
        container.append(interim)
    
    return pd.concat(container)
```

```python
def processing(wp_top10):

    ER_wp_1 = equality_ratio(wp_top10, denominator=1)
    ER_wp_1['N_ties'] = 1

    ER_wp_2 = equality_ratio(wp_top10, denominator=2)
    ER_wp_2['N_ties'] = 2

    ER_wp_3 = equality_ratio(wp_top10, denominator=3)
    ER_wp_3['N_ties'] = 3

    ER_wp = pd.concat([ER_wp_1, ER_wp_2, ER_wp_3])

    ER_wp_stats = cl.util.utils.ci_transfer(ER_wp, on=['Rank', 'N_ties'], target='equality_ratio')
    
    return ER_wp_stats
```

```python
wp_stats = processing(wp_top10)
```

```python
BZR_top10 = pd.read_csv('final/extra/BZR_reciprocal_30_no_call_common_top10_details.csv')
BZR_stats = processing(BZR_top10)
```

```python
# CATEGORY_COLUMN = 'category'
DATASET_COLUMN = 'dataset'
USERID_X_COLUMN = 'userid_x'
USERID_Y_COLUMN = 'userid_y'

def jaccard_similarity(list1, list2):
    s1 = set(list1)
    s2 = set(list2)
    return len(s1.intersection(s2)) / len(s1.union(s2))


def plot_similarity(category, data, userlist, CATEGORY_COLUMN,
                    local=False,
                    mode='talk', style="whitegrid", l=10, w=6):
    n_cat = len(category)
    
    
    if n_cat > 1:
        sns.set_context(mode)
        sns.set_style(style)
        fig, ax = plt.subplots(figsize=(l, w))

        if local:
            # Compute local Jaccard similarity
            js_list = []
            for i in range(n_cat):
                cat1 = category[i]
                for j in range(i + 1, n_cat):
                    cat2 = category[j]
                    item_name = str(cat1) + ' vs ' + str(cat2)
                    for user in userlist:
                        alter_cat1 = \
                            data[(data[CATEGORY_COLUMN] == cat1) & (data[USERID_X_COLUMN] == user)][
                                USERID_Y_COLUMN].tolist()
                        alter_cat2 = \
                            data[(data[CATEGORY_COLUMN] == cat2) & (data[USERID_X_COLUMN] == user)][
                                USERID_Y_COLUMN].tolist()
                        js = jaccard_similarity(alter_cat1, alter_cat2)
                        js_list.append([item_name, js])

            result = pd.DataFrame(js_list, columns=['item_name', 'js_value'])

            sns.boxplot(data=result, x='item_name', y='js_value', ax=ax)
            ax.set_xlabel('')
            ax.set_ylabel('Local Jaccard Similarity')
            if len(result['item_name'].unique()) > 1:
                ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

        else:
            # Compute global Jaccard similarity
            js_list = []
            for cat1 in category:
                for cat2 in category:
                    alter_cat1 = data[data[CATEGORY_COLUMN] == cat1][USERID_Y_COLUMN].tolist()
                    alter_cat2 = data[data[CATEGORY_COLUMN] == cat2][USERID_Y_COLUMN].tolist()
                    js = jaccard_similarity(alter_cat1, alter_cat2)
                    js_list.append(js)
            result = np.array(js_list).reshape(n_cat, n_cat)
            result = pd.DataFrame(result, columns=category, index=category)

            sns.heatmap(result, cmap="YlGnBu", ax=ax,
                        linewidths=.5, annot=True)
            # ax.xaxis.tick_top() # x axis on top
            ax.xaxis.set_label_position('bottom')
            ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
            ax.set_yticklabels(ax.get_yticklabels(), rotation=0, ha='right')

        return fig
    else:
        return 'Must have two networks!'
```
