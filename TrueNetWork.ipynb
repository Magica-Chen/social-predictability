# True Friendship network (ranking by meetups and previous locations)

```python
import numpy as np
import pandas as pd
%load_ext autoreload
%autoreload 2
import meetup_strategy as ms
import time 
import pickle
```

```python
pickle_in = open("n_previous_placeidT.pickle", "rb")
placeidT = pickle.load(pickle_in)
pickle_in.close()
valid_user = list(placeidT.keys())
```

## Meetupers and True Friendship filter

```python
df_friend=pd.read_csv('data/weeplace_friends.csv')
df_friend = df_friend[(df_friend['userid1'].isin(valid_user)) & (df_friend['userid2'].isin(valid_user))]
```

```python
df_friend['userid1'].nunique()
```

```python
user_meetup = pd.read_csv('n_previous_total_meetup.csv')
user_meetup
```

```python
user_meetup['userid_x'].nunique()
```

```python
share_ego = list(set(df_friend['userid1'].tolist()) & set(user_meetup['userid_x'].tolist()) )
```

```python
share_friends = df_friend[df_friend['userid1'].isin(share_ego)]
share_friends = share_friends.rename({'userid1': 'userid_x', 'userid2': 'userid_y'}, axis=1)
share_friends
```

```python
share_meetupers = user_meetup[user_meetup['userid_x'].isin(share_ego)]
share_meetupers
```

```python
share_friend_meetup = share_friends.merge(share_meetupers, how='left', on=['userid_x', 'userid_y'])[['userid_x', 
                                                                                                    'userid_y',
                                                                                                    'meetup']].fillna(0.0)
share_friend_meetup
```

## Generate more details about true network

```python
TrueFriendNet = ms.FriendNetwork('data/weeplace_checkins.csv', share_friends, placeidT=placeidT)
```

```python
TrueFriendNet.ego_info(verbose=True, filesave=True)
```

```python
filename = 'True-Friendship-details.csv'
TrueFriendNet.total_meetup.to_csv(filename, index=False)
```

```python
true_meetup = share_friend_meetup.merge(TrueFriendNet.total_meetup, 
                                        how='left', 
                                        on=['userid_x', 
                                            'userid_y']).sort_values(['userid_x', 
                                                                      'meetup', 
                                                                      'N_previous'], 
                                                                     ascending=[True, False, False])
true_meetup = true_meetup[['userid_x', 'userid_y', 'meetup', 'N_previous', 'group']]
```
