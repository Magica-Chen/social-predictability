```python
df_ent=pd.read_csv('weeplace_150_ent.csv')
```

```python
df_ent
```

```python
# pdata_ent = pdata.copy()
# pdata_ent['LZ_entropy'] = pdata.apply(lambda row: ent_all[row.userid], axis=1)
pdata_ent = pdata.merge(df_ent, how='left', on='userid')
```

```python
placeid_count = pdata.groupby('placeid').agg({"userid": "nunique"}).reset_index()
```

At least 50 distinct users visited this location

```python
df_valid_placeid = placeid_count[placeid_count['userid'] >=50]
```

```python
valid_placeid = df_valid_placeid['placeid'].tolist()
```

```python
len(valid_placeid)
```

```python
pdata_stats = pdata_ent[pdata_ent['placeid'].isin(valid_placeid)].groupby('placeid')['LZ_entropy'].agg(['mean', 
                                                                                                         'median',
                                                                                                        'std',
                                                                                                         'count', 
                                                                                                         'nunique']).reset_index()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pdata_stats
```

```python
sns.set_context('talk')
fig, ax= plt.subplots()

sns.distplot(pdata_stats['mean'], ax=ax, label='Mean')
sns.distplot(pdata_stats['median'], ax=ax, label="Median")

ax.legend(loc='upper left')
```

```python
df_checkin_full = df_checkin.copy()
df_checkin_full['category'], df_checkin_full['sub_category'] = df_checkin['category'].str.split(':', 1).str
```

```python
df_checkin_full = df_checkin_full.replace({'category' : { 'Great Outdoors' : 'Parks & Outdoors', 
                                       'Nightlife' : 'Nightlife Spots', 
                                       'Travel' : 'Travel Spots',
                                      'Colleges & Universities': 'College & Education',
                                      'Home / Work / Other': 'Homes, Work, Others'}}).dropna()
```

```python
list(set(df_checkin_full['category'].tolist()))
```

```python
pdata_stats_cat = pdata_stats.merge(df_checkin_full[['placeid', 'category']], 
                                    how='left', 
                                    on=['placeid']).drop_duplicates(subset=None, keep='first')
```

```python
# fig, ax = plt.subplots(figsize=(10, 8))
sns.displot(pdata_stats_cat, x="median", hue="category", kind="kde", height=6, aspect=2)
```

```python
sns.displot(pdata_stats_cat, x="mean", hue="category", kind="kde", height=6,aspect=2)
```

```python
sns.displot(pdata_stats_cat, x="std", hue="category", kind="kde", height=6,aspect=2)
```

```python
fig, ax= plt.subplots(figsize=(12,8))
g = sns.violinplot(x="category", y="mean",data=pdata_stats_cat)
g.set_xticklabels(g.get_xticklabels(), rotation=60)
```
