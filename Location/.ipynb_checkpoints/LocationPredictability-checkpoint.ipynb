```python
import numpy as np
import pandas as pd
import colocationship as cl
import copy

%load_ext autoreload
%autoreload 2
```

# Initial Processing: Consider the users with more than 150 check-ins

```python
df_checkin = pd.read_csv("../data/weeplace_checkins.csv")
co_locationship = cl.Co_Locationship(df_checkin, mins_records=150, missing='-')
```

```python
placeidT = cl.util.read_object("../temp_data/wp-150/weeplace_placeidT_150.pickle")
co_locationship.placeidT = placeidT
```

```python
print('There are', co_locationship.pdata['userid'].count(), 'valid check-ins records, \n'
      'There are', co_locationship.pdata['userid'].nunique(), 'valid users, \n'
      'There are', co_locationship.pdata['placeid'].nunique(), 'distinct placeid. \n'
     )
```

```python
pdata = co_locationship.pdata
```

```python
def get_ent_Pi(k, v):
    seq = v['placeid'].tolist()
    N = len(set(seq))
    S = cl.util.LZ_entropy(seq)
    Pi = cl.util.getPredictability(N, S)
    return [k, S, Pi]
```

```python
df_ent_Pi = [get_ent_Pi(k, v) for (k, v) in placeidT.items()]
df_ent_Pi = pd.DataFrame(df_ent_Pi, columns=['userid', 'LZ_entropy', 'Pi'])
```

```python
df_ent_Pi
```

```python
A = df_checkin[['placeid', 'lat', 'lon']].drop_duplicates()
```

```python
df_checkin
```

```python
pdata_ent_Pi = pdata.merge(df_ent_Pi, how='left', on='userid')
```

```python
placeid_count = pdata.groupby('placeid').agg({"userid": "nunique"}).reset_index()
```

At least 50 distinct users visited this location

```python
df_valid_placeid = placeid_count[placeid_count['userid'] >=50]
```

```python
valid_placeid = df_valid_placeid['placeid'].tolist()
```

```python
len(valid_placeid)
```

```python
pdata_ent_Pi
```

```python
pdata_stats_ent = pdata_ent_Pi[pdata_ent_Pi['placeid'].isin(valid_placeid)].groupby('placeid')['LZ_entropy'].agg(['mean', 
                                                                                                         'median',
                                                                                                        'std',
                                                                                                         'count', 
                                                                                                         'nunique']).reset_index()
```

```python
pdata_stats_Pi = pdata_ent_Pi[pdata_ent_Pi['placeid'].isin(valid_placeid)].groupby('placeid')['Pi'].agg(['mean', 
                                                                                                         'median',                                                                                                         
                                                                                                         'std',
                                                                                                         'count',
                                                                                                        'nunique']).reset_index()
```

```python
pdata_stats_full = pdata_ent_Pi[pdata_ent_Pi['placeid'].isin(valid_placeid)].merge(pdata_stats_ent, 
                                                                how='left', 
                                                                on='placeid').merge(pdata_stats_Pi, 
                                                                                    how='left',
                                                                                    on = 'placeid',
                                                                                    suffixes=('_ent', '_Pi'))
```

```python
pdata_stats_ent['target'] = 'Entropy'
pdata_stats_Pi['target'] = 'Pi'
pdata_stats = pd.concat([pdata_stats_ent, pdata_stats_Pi])
```
