# Gender aspect of meetupers (Weak Confidence)

```python
user_meetup_AM = pd.read_csv('results/user-meetup-info-M_AM.csv')
user_meetup_AF = pd.read_csv('results/user-meetup-info-F_AF.csv')

user_meetup_G = pd.read_csv('results/user-meetup-info-G.csv')
```

```python
user_meetup_AM['Gender_Guesser_y'] = 'M'
user_meetup_AF['Gender_Guesser_y'] = 'F'
```

we consider both 'M' and 'AM' as male, and both 'F' and 'AF' as female

```python
def str_replace(x):
    if x == 'AM':
        return 'M'
    elif x == 'AF':
        return 'F'
    else:
        return x
```

```python
user_meetup_G['Gender_Guesser_y'] = user_meetup_G.apply(lambda row: str_replace(row.Gender_Guesser_y),axis=1)
```

## Brief introduction of current dataset

```python
shared_egolist = list(set(user_meetup_AM['userid_x'].tolist()) & 
                      set(user_meetup_AF['userid_x'].tolist()) & 
                      set(user_meetup_G['userid_x'].tolist())) 
```

```python
len(shared_egolist)
```

It looks we can have more valid users.

```python
user_stats_M = user_meetup_AM[user_meetup_AM['userid_x'].isin(shared_egolist)]
user_stats_F = user_meetup_AF[user_meetup_AF['userid_x'].isin(shared_egolist)]
user_stats_MF = user_meetup_G[user_meetup_G['userid_x'].isin(shared_egolist)] 
```

```python
print(len(set(user_stats_M['userid_y'].tolist())), 
     len(set(user_stats_F['userid_y'].tolist())),
     len(set(user_stats_MF['userid_y'].tolist())))
```

In this weak confidence dataset, for all the shared users, there are **6171** male meetupers and **1665** female meetupers, totally, **7836** meetupers

```python
user_stats = pd.concat([user_stats_M, user_stats_F])
vs.hist_entropy_gender(user_stats, user_stats_MF)
plt.show()
vs.hist_pred_gender(user_stats, user_stats_MF)
plt.show()
```

## Compare entropy and predictability with only male meetupers and female meetupers

```python
vs.num_point_plot_gender(user_stats, name='predictability', user_stats_mixed=user_stats_MF, mode='paper', threshold=150)
```
